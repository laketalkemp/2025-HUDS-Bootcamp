## Concatenating all of the files into one dataframe.

```python
# Initialize dataframes as an empty list to store DataFrames
dataframes = []

# Loop through the list of full filenames and read each into a pandas DataFrame
# Use 'filenames' (the output from the previous step) instead of 'filename'
for file_path in filenames:
    # Check if the file actually exists before trying to read it
    if not os.path.exists(file_path):
        print(f"Warning: File not found at {file_path}. Skipping.")
        continue # Skip to the next file if this one doesn't exist

    try:
        df = pd.read_stata(file_path)
        if not df.empty: # Only append if the DataFrame is not empty
            dataframes.append(df)
        else:
            print(f"Warning: Stata file {file_path} read successfully but is empty. Skipping.")
    except Exception as e:
        print(f"Error reading Stata file {file_path}: {e}. Skipping.")

# Concatenate the dataframes. This will only run if 'dataframes' is not empty.
if dataframes:
    # If concatenating along axis=1 (columns), ensure indices are aligned or use ignore_index=True
    # Using axis=1 usually means you expect columns to be added, implying
    # all dataframes should have the same number of rows or a common index.
    # If the number of rows can vary and you want to fill missing values with NaN,
    # you might need to reindex or ensure consistent indices.
    # For a simple column-wise concatenation where indices might not perfectly match,
    # consider pd.concat(dataframes, axis=1, join='outer')
    # or pd.concat(dataframes, axis=1, join='inner') depending on your needs.
    # Here, 'outer' join is used as a common default.
    ghana_dataframe = pd.concat(dataframes, axis=1, join='outer')
    print("\nSuccessfully concatenated DataFrames. Head of the combined DataFrame:")
    print(ghana_dataframe.head())
    print(f"Combined DataFrame shape: {ghana_dataframe.shape}")
else:
    print("\nNo DataFrames were successfully loaded. 'ghana_dataframe' was not created.")
    print("Please check the warnings above for details on files that were skipped due to errors or being empty.")
```
